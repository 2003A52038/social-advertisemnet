```python
import seaborn as sns
import matplotlib.pyplot as plt
def get_heatmap(df):
    plt.figure(figsize=(6,4))
    sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
    plt.tight_layout()
    return plt.show()
get_heatmap(df)
```
